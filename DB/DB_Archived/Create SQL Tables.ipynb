```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('df_Combined.csv')
```

```python
data.fillna(0, inplace = True)
```

```python
data
```

```python
users_table = pd.DataFrame(data['Name'].copy())
users_table.reset_index(inplace = True)
users_table.columns = ['user_id', 'name']
users_table
```

```python
locations = pd.DataFrame(data.columns[1:])
locations_table = pd.DataFrame(locations)
locations_table.reset_index(inplace = True)
locations_table.columns = ['location_id', 'location_name']
locations_table
```

```python
#Create ratings table:

reformat_data = data.copy()
reformat_data.drop('Name', axis = 1, inplace = True)
reformat_data.columns = np.arange(0, len(reformat_data.columns))

reformat_data
```

```python
ratings_data = []

for user in users_table['user_id']:
    user_ratings = reformat_data.loc[user]
    user_ratings = user_ratings[user_ratings != 0]
    for location, rating in enumerate(user_ratings): 
        ratings_data.append([user, user_ratings.index[location], rating]) 

ratings_table = pd.DataFrame(ratings_data)        
ratings_table.columns = ['user_id', 'location_id', 'rating']

ratings_table
```

```python
#Data Dump:

users_table.to_csv('users_table.csv')
locations_table.to_csv('locations_table.csv')
ratings_table.to_csv('ratings.csv')

```
