```python
html_text = requests.get("https://www.tripadvisor.com/Restaurants-g60713-San_Francisco_California.html")
soup = bs.BeautifulSoup(html_text.content, features='html.parser') 

# getting the number of pages that contain restaurants
page_numbers = soup.find_all(class_ = "pageNum taLnk")
#page_numbers = page_numbers.split("\n")
page_numbers
numbers = []
for number in page_numbers:
    numbers.append(number.text.strip())
number_pages = int(numbers[len(numbers)-1]) # number of pages with restaurants, we have to crawl
```

```python
### changing the code to run this for multiple pages
## scraping multiple pages, getting the restaurant URLs
t0 = time.time()
urls_restaurants = []
pre_url = "https://www.tripadvisor.com"
for i in range(0, number_pages):     # if I run this I get all the restaurant urls
#for i in range(0, 1):
    
    url = "https://www.tripadvisor.com/RestaurantSearch-g60713-oa" + str(i*30) + "-San_Francisco_California.html"
    print(url)
    html_text = requests.get(url)
    soup = bs.BeautifulSoup(html_text.content, features='html.parser') 
    all_restaurants = soup.find_all(class_ = "_1llCuDZj")
    for r in all_restaurants:
        url = r.find("a").get('href') 
        urls_restaurants.append(pre_url + url)
t1 = time.time()
total_time = t1-t0
```

```python
t00 = time.time()


#starting with the read in datafram
full_df = read_in
#full_df = pd.DataFrame(columns = ["User_Name"])


## code to scrape the reviews of the restaurant
a = 1
for url in urls_restaurants:
    """if a >3:
        break
    a+=1"""
    print(url)
    html_text = requests.get(url)
    soup = bs.BeautifulSoup(html_text.content, features='html.parser') 
    restaurant_name = soup.find("h1").text
    if restaurant_name in restaurant_dict:
        print("skipping ", restaurant_name)
        continue
    
    
    
    ### find number of pages for the reviews
    page_object = soup.find(class_="pageNum last cx_brand_refresh_phase2")
    if page_object:
        number_review_pages = int(page_object.text)
    else:
        number_review_pages = 1
    print(number_review_pages)

    ## all the urls/pages for the review pages
    restaurant_urls = []
    
    ## reducing number of reviwes we look at!!
    #if number_review_pages >50:
    #    number_review_pages = 50
        
    
    for i in range (0,number_review_pages):
        url_parts = url.split("Reviews")
        new_url = url_parts[0]+"Reviews-or"+ str(i*10)+ url_parts[1]
        restaurant_urls.append(new_url)
    #print(restaurant_urls)
    
    
    ### add loop over all the pages of reviews
    
    ratings = []
    user_names = []
    for rest_url in restaurant_urls:
        #print(rest_url)
        html_text = requests.get(rest_url)
        soup = bs.BeautifulSoup(html_text.content, features='html.parser') 
        reviews_text = soup.find(id="REVIEWS").find(class_="ppr_rup ppr_priv_location_reviews_list_resp").find(class_ = "listContainer hide-more-mobile")
        review_container = reviews_text.find_all(class_="review-container")
        for div in review_container:
            #print(div.find(class_="memberOverlayLink clickable").text)
            user_object = div.find(class_="memberOverlayLink clickable")
            if not user_object:
                continue
            user_name = div.find(class_="memberOverlayLink clickable").text.strip()
            rating = float(div.find(class_="ui_bubble_rating").get("class")[1].split("_")[1])/10
            user_names.append(user_name)
            ratings.append(rating)
    #print(ratings)
    #print(user_names)
    data = {"User_Name": user_names, restaurant_name: ratings}
    print(data)
    df_new = pd.DataFrame(data, columns = ["User_Name", restaurant_name ])
    full_df = full_df.merge(df_new, on='User_Name', how='outer')
    full_df.to_csv('Iteration1.csv', index = None)
    #test_df = test_df.merge(df_new, on='User_Name', how='inner')
#full_df.to_csv('test_duplicates.csv', index = None)
t11 = time.time()
total_time2 = t11-t00
print(total_time2)
```
